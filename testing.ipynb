```python
from tensorflow import keras
import cv2
import numpy as np
from keras.preprocessing import image
from PIL import Image as im

from keras.models import Sequential
from keras.layers import Dense,Conv2D,Dropout,MaxPooling2D,Flatten
```

```python
model = keras.models.load_model('drowsy.h5')
```

```python
def testImageUpload(test_img,model):
    test_img = test_img.resize((100,100))
    x = image.img_to_array(test_img)
    x = x/255
    x = np.expand_dims(x,axis = 0)
    images = np.vstack([x])
    classes = model.predict(images, batch_size = 10)
    print("ClosedEye :",classes[0][0])
    print("OpenEye :",classes[0][1])
    

def faceDetection(test_img):
    gray_img = cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)
    faceHaarCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    faces = faceHaarCascade.detectMultiScale(gray_img)

    return faces, gray_img

cap = cv2.VideoCapture(0)
cv2.namedWindow("test")

while True:
    ret, test_img = cap.read()
    faces_rect, gray_img = faceDetection(test_img)
    if len(faces_rect) != 1:
        continue
        
    x,y,w,h = faces_rect[0]
    roi_colour = test_img[y:w+h,x:x+h]
    data = im.fromarray(roi_colour)
    testImageUpload(data,model)
    
    if cv2.waitKey(10) == ord('q'):#wait until 'q' key is pressed
        break
        

cap.release()
cv2.destroyAllWindows()
    

# cv2.imshow('roi_colour',roi_colour)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

```
